```python
import pandas as pd

pacjenci = pd.DataFrame([
    {
        "imie": "Jan",
        "nazwisko": "Kowalski",
        "pesel": "90010112345",
        "telefon": "500 600 700",
        "adres": "ul. Prosta 12, Warszawa",
        "alergie": "brak",
        "ostatnia_wizyta": "12.03.2026 - Internista"
    }
])

pacjenci
```

```python
def wyszukaj_pacjenta(pesel):
    wynik = pacjenci[pacjenci["pesel"] == pesel]
    if wynik.empty:
        return "Nie znaleziono pacjenta"
    return wynik

wyszukaj_pacjenta("90010112345")
```
